```python
def extract_timeline_as_df(time_line_list):
    columns = set()
    allowed_types = [str, int]
    tweets_data = []
    for status in time_line_list:
        # print(status.author.screen_name)
        status_dict = dict(vars(status)) # The vars() function returns the __dic__ attribute of an object.
        # print(status_dict)
        # print(status.text)
        keys = status_dict.keys()
        # print(keys)
        single_tweet_data = {"user": status.user.screen_name, "author": status.author.screen_name}
        for k in keys:
            try:
                # print(k)
                # print(status_dict[k])
                v_type = type(status_dict[k])
            except:
                v_type = None
            # print(v_type)
            if v_type != None:
                if v_type in allowed_types:
                    # print(v_type)
                    # print(status_dict[k])
                    single_tweet_data[k] = status_dict[k]
                    # print(single_tweet_data[k])
                    # print(k)
                    columns.add(k)
        tweets_data.append(single_tweet_data)


    header_cols = list(columns)
    header_cols.append("user")
    header_cols.append('author')
    # print(header_cols)
    # print(tweets_data)
    df = pd.DataFrame(tweets_data, columns=header_cols)
    return df
```

```python
my_timeline = api.home_timeline()
```

```python
df2 = extract_timeline_as_df(my_timeline)
df2.head()
```
